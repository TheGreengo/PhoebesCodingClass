## If statements

In code, there are certain times where we might have several possible situations we need to handle with a single block of code. This is handled with conditionals.

The simplest conditional is an "if" statement. An if statement starts with the keyword "if" and then a boolean expression that is either true or false followed by a colon and some indented code. If the statement is true, the indented code following it is executed.

```python
if True:
  print("thing")

if False:
  print("what I put here doesn't matter 'cause it's not going to print")
```

Any code that is not in the indented lines following the colon will execute regardless of the truth of the expresion.

```python
if False:
  print("will not print")
print("this part prints regardless")
```

## Conditional Statements

Now, it's obviously very limited if we just type a boolean or place a variable with of boolean type in our "if" statements. If that were the case, they would honestly be quite close to useless.

In python, there are certain operations that return boolean values. For example, one can compare values with "<" or ">" symbols. In the examples below, notice that these statements return a boolean value corresponding to whether or not the statement is accurate.

```python
print(4 > 1)
```

```python
print(30 < 25)
```

Unfortunately, "=" is reserved for assigning values not comparing them, so if we used this, it would change the value of any variables in the first spot instead.

```python
print(4 == 2 + 2)
```

```python
type(4)
```

```python
type(4.0)
```

```python
print(type(4.0) == type(4))
```

Conversely, one can check if two values are different with the "!=" operator.

```python
print(1 != 2)
```

```python
word = "thing"
print(word != "thing")
```

For "greater than or equals to" one uses ">=" and similarly "<=" for "less than or equal to".

```python
print(3 >= 3)
```

```python
print(2 <= 4)
```

## Other Conditional Evaluators

### Else
"If" statements can be combined with an "else" statement. If the statement in the "if" statement evaluates to false, then any indented code following the "else" is evaluated.

If the statement is true, then the indented code after the if statement is run, but not the indented code after the "else" statement.

```python
if (3 < 4):
  print("this will print")
else:
  print("this won't")

print("since this isn't indented, it will run regardless")
```

```python
if (3 > 4):
  print("this won't print")
else:
  print("this will")

print("since this isn't indented, it will run regardless")
```

## Elif
"Elif" is an abreviation for "Else-If". After an "if" statement, one can include an "elif" with another conditional statement. If statement after the "if" is NOT true, AND the conditional statement after the elif IS true, then the indented code after the "elif" is run.

```python
value = 5

if value > 7:
  print("don't worry about it")
elif value < 6:
  print("this happened")
```

```python
value = 5

if value > 4:
  print("this prints")
elif value < 6:
  print("even though this statement is true, since the original statement is as well, only that part executes")
```

To create a more advanced conditional, one can use an "if" statement together with an "else" statement and as many "elif" statements as you want.

In this form, the statements evaluate from top to bottom in order. As soon as one "elif" statement is found to be true, all others after it are ignored.

```python
value = 945395

if value > 50:
  print("this is a really large number")
elif value > 25:
  print("this is a decently big number")
elif value > 5:
  print("this is a relatively small number")
else:
  print("this is a small number")
```

## "and" and "or" and "not"

To make sophisticated conditional statements, one can combine multiple statements together with "and" or "or".

Statements joined with "and" only evaluate as true if all the statements individually evaluate to being true.

Statements joined together with "or" evaluate together as true if ANY of them are true.

```python
if True and True:
  print("lolz")
```

```python
if True or False:
  print("yeet")
```

Additionally, one can preface a truth statement with "not" and the statement will return the opposite.

```python
print(not 5 < 3)
```

```python
print(not 5 < 309854)
```

## Embedded Conditionals

Inception time baby.

You can put if/else statements inside the indented block following another if/else statement with a further indented block after it. As you might expect, if the "parent" if statement is true, then the if/else statement within it will evaluate.

In theory, there's no limit to how many levels one can get.

```python
num = 38

if num < 50:
  if num < 25:
    if num < 12:
      print("really small")
    else:
      print("kinda small")
  else:
    if num < 37:
      print("kinda big")
    else:
      print("really big")
else:
  print("too big")
```
