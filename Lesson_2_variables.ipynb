# Types

Your computer needs to store everything as 1s and 0s. As such, it needs to know how to encode different kinds of input. This means that any variable which is being used to store a value, also has a type for the value it’s storing. There are a lot of different types that python can work with, but we’ll go over some of the most useful and basic right now and work on some of the more complicated ones later.

### int
An "int" is a whole number (no decimal or fractions) but can be positive or negative. They can also have really big values.

```python
this_is_an_int = 0
also_an_int = 23932483
still_intish = -12349700
it_should_be_spelled_ent = -69420
```

### string
A string is a set of characters. To create a string, values must be enclosed in either '' or "". Strings can contain most any character you can type with your keyboard. We'll talk more about strings in the next lesson.

```python
string = "string beans"
urbtt = "your butt"
single ='like an f&*^ing pringle'
you = "totally won this one I'm losing my mind"
python = "named Cody like 'coding'"
```

### float
The name here is weird, and in order for it to make sense you have to get a lot deeper into computer science than is healthy, but a float is essentially just a decimal number.

```python
floatiness = 420.69

#And they can be negative too

more_floating = -928734987423.44

#Somewhat confusingly, this is still a float not an int

weirdness = 69.0

computer = 5700815.69
```

### bool
Now, if you remember the story about the reason two mathematicians can’t get married, this is named after him, George Boole. His focus in math was on the different mathy things you can do in systems with only two number, so a bool takes on only one of two values “True” or “False”

```python
his_death_avoidable = True
it_is_smart_that_bools_are_capitilized_in_python = False
the_ultimate_wet_blanket = True
```

### NoneType
This one isn't useful starting out, but "None" is used as a placeholder sometimes and is important since it allocates a variable but only reserves the minimum amount of storage.

```python
empty = None
catch_all = None
```

### Implicitness

Python is what’s known as an “implicitly-typed” language. This just means that when you create a variable, your computer will work on its own to figure out what the type is of the variable you just created.

It is important however to note that just because you don’t ever mention the type doesn’t mean that it’s not there.

One useful trick is the fact that if you put a variable in the “type()” function, it will automatically return the type of that variable.


```python
test = None
print(type(test))
```

### Changing Type
One super useful (and borderline anarchy-inducing) is that you can even change the type of a variable. Try running the following three blocks in succession.

```python
changing = "a string"
print(type(changing))
print(type(7))
print(type("Python is cute"))
```

```python
changing = 0
print(type(changing))
```

```python
changing = True
print(type(changing))

changing = 3750
print(type(changing))
```
