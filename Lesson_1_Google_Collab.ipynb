### Markdown

When trying to add text or after double clicking on a text block, you might notice that the cell splits in two halves. If you haven't come across that yet, try double clicking this cell now.

Weird huh?

This is because the text chunks are written in a text formatting language called markdown. Since you are a language nerd like myself, I'm going to include some etymology here for you in order to explain what this is:

Back in ye olde days of yore, printers would prepare book manuscripts by "marking up" a manuscript with instructions in a sort of special annotation detailing what fonts, sizes, and features (italics, underlines, bold, etc.) should be used when preparing the printing presses.

When the ENQUIRE (precursor to the internet) was invented, the researchers would share documents by sending raw .txt files. This meant no highlighting, no italics, no font size, just letters and numbers. Since ENQUIRE was used by researchers sharing technical files, one man named Tim Berners-Lee created a "hyper-text markup language" or HTML for short, to do the same "markup" preperation that old printers would do, so that fellow researchers could better style the documents they sent to each other, increasing communication.

As the internet developed, many people created their own markup languages, and they were typically quite sophisticated.

In 2004, a team made "Markdown" a markup language aimed to do more simple text formatting with a humurous name to play on the term "markup".

You absolutely don't need to use markdown when adding text to a collab file, you can totally just type it in or use the formatting options that google adds for you (visible on the top left when editing a code block), but when you click on it and see funny shapes, you might wonder why it looks so weird so I figured I'd explain what it is.

The basic things (the only things I ever use) are:

-----

Start a line with a hashtag '#' for a header, a double hashtag '##' for a smaller header, or a triple '###' for an even smaller header.

# Header
## Smaller Header
### Even Smaller Header

Start a line with a '-' or '*' for a list.

- list item
* also a list item

use an extra space before the '-' or '*' for a list within a list

- list
 - list within list

Or with a number and a period for a numbered list.

1. Thing one
2. Thing two

There's more you can do but it's not really that useful.
