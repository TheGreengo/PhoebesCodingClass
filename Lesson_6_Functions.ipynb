# Backstory

Computer science is a field that developed as the results of groundbreaking inovations by engineers and mathematicians. As both fields fought for dominance in the new domain of academia, mathematicians won in all of the wrong ways.

You see, originally, some people referred to programs as routines. The same way you lay out an exercise or work routine and then go through the steps one by one until your done, you make a list of tasks for the computer to do, and it goes through the instructions one by one until it's done.

These engineers realized that many times in programs, there were sets of instructions that needed to be repeated at verious different points in the "routine" (not just multiple times in a row like a loop). They referred to these instruction sets as "sub-routines" sets of instructions that could be treated like a routine that was carried out itself multiple times in the course of a larger routine.

Mathematicians however came along. They realized that sub-routines behaved a little bit like functions in math (though they are not functions by math's standards) and named them functions.

And they won.

So we're going to learn about functions.

# Functions

This is where functions would be helpful. A function is essentially a named block of code. Everytime that you invoke that name, the same block of code executes. This allows you to repeat a particular block of code at different points in a larger program.

In python, you create a function using the `def` keyword followed by the name of the function, ( ), and then a colon. Any indented code imediately following is the body of the function, and will be executed whenever you later write the function name followed by ( ).

```python
def yourMom():
  print("Your mom")
  print("is ")
```

```python
yourMom("cool")
yourMom()
```

Functions can include if/else statements, loops, and even other functions in their bodies and as a result often can contain multiple levels of indentation.

```python
def lotsOfStuff():
  for i in range(10):
    if (i % 2) == 0:
      yourMom()
    else:
      print("Your butt")
```

## Arguments

There are times where we might want to repeat a particular process, but with some slight variation. You've already encountered this a lot in this course with the `print()` function. The function does the same thing every time, prints something to the output of your computer. But what it prints varies depending on what you put in.

Things that go into functions are called arguments. When defining a function, you can place variable names in the parentheses to indicate that when the function is called, that many values will be put in the parentheses. These values can be accessed within the function body, using the variable names given in the function definition.

For example, we can write a function that allows the user to give a number of repitions and a string they'd like printed out, then the function will print it that many times.

```python
def add(num1, num2):
    num1 = num1 * 2
    num = num1 + num2
    print(f"the sum of {num1} and {num2} is {num}")
```

```python
def tempConversion(in_temp, FtoC):
    if FtoC:
        out_temp = (in_temp - 32) * (5 / 9)
    else:
        out_temp = (in_temp * 1.8) + 32
    print(f"Your mom is {out_temp :.2f}")
```

```python
def rec(num1, num2):
    return num1 * num2
```

```python
def coob(num1, num2, num3):
    temp = rec(num1, num2)
    coob = rec(temp, num3)
    print(coob)
```

```python
def circ(rad):
    pi = 3.14159265358
    return pi * (rad**2)

def cylinder(rad, height):
    return circ(rad) * height

def cone(rad, height):
    return cylinder(rad, height) * (1/3)
```

```python
print(cone(2, 8))
```

```python
coob(2,3,4)
```

```python
tempConversion(69, True)
tempConversion(69, False)
```

```python
add(5, 4)
```

### Default Values
Functions can also be programmed with default values so that if certain parameters aren't specified, the function will fill in the default value when executing the function.

For example, below we have a temperature converter which can convert temperatures from fahrenheit to celsius or from celsius to fahrenheit. In order to let this function handle both we have a boolean parameter "f_to_c". If true, then the function converts from fahrenheit to celsius, if false it converts from celsius to fahrenheit.

```python
def tempConvert(in_temp, FtoC):
    if FtoC:
        out_temp = (in_temp - 32) * (5 / 9)
    else:
        out_temp = (in_temp * 1.8) + 32
    print(f"Your mom is {out_temp :.2f}")
```

```python
tempConvert(75, True)
tempConvert(45, False)
```

We may want people to be able to just put a number into the function. If we think more people will be using this software to convert from f to c then we might set the default value of "f_to_c" to be true so that people can just put in a number and convert from fahrenheit to celsius.

If the second parameter is listed, then the function will use it's value, but if only one is given, it automatically assigns `f_to_c = True`.

```python
def tempConvert(in_temp, FtoC = False):
  if FtoC:
      out_temp = (in_temp - 32) * (5 / 9)
  else:
      out_temp = (in_temp * 1.8) + 32
  print(f"Your mom is {out_temp :.2f}")
```

```python
tempConvert(45)
```

## Return Values

One of the most useful features of functions is that they can give a value back at the end. This is called a return statement because you use the keyword `return` to indicate that the function will give back a certain value.

In the example below, your computer will look at the right side of the equals sign, and replace it with what `concat("your","butt")` returns.


```python
def example(num):
    num *= num
    return num
```

```python
def concat(word1, word2):
  return word1 + " " + word2
```

```python
val = concat("thing", "thang")
print(val)
dog = input("your dog")
print(dog)
```

```python
thing = concat("your","butt")

print(thing)
```

### Ending

One thing that is important to understand about functions is that they stop after encountering an instance of the `return` keyword.

Some functions have multiple return statements, but this is if there are if/else statements and any one call of the function will only go through one return statement.

For example, the following function can return "Even" or "Odd", so, when we write it, it contains two return statements, but depending on the input, it will only actually reach one of them. If the number is even, the first if statement evaluates to true and the function returns, ignoring the rest of the code. If the number is odd, the first indented codeblock is ignored, and the final return statement is processed.

```python
def isEven(num):
  if num % 2 == 0:
    return "Even"
  return "Odd"
```

```python
thang = isEven(5)
print(thang)
```
