# For Loops
Harken back to our example yesterday, where we imagined coding up one's daily activities. If you wrote it out with the skills we've gone over so far, you would probably find yourself doing a lot of copying and pasting. If you washed dishes, you would find yourself writing some code like

`wash(plate1)`

`wash(plate2)`

`wash(bowl1)`

`wash(bowl2)`

And so on for each of your dishes. Similarly, you might find yourself writing something like 

`read(email1)`

`read(email2)`

`read(email3)`

and so on for each one of your emails. Processes like these, where we repeat a function or procedure for each element in a list (i.e. each dish in the sink, or each email in your inbox), that very early on in the history of programming, coders developed ways to "loop" through the elements of a list to repeat a process. These procedures are called "for" loops, or "for each" loops, and are a staple of modern programming.

### For Loop Syntax
Let's take our dishwashing for example, since it's your favorite thing in the whole world to do. Below is a list of our dishes that we want to wash and a loop that would go through each item in the list and print `washing <dish_name>` and then `<dish_name> has been cleaned!`.

Note that the syntax boils down to `for <placeholder_name> in <list_name>:` and then an indented block of code. In the block of code, we can use the placeholder name where we want the values of each of our dishes to be read in.

```python
dishes = [ "plate", "plate", "bowl", "bowl", "cup", "big bowl"]

for dish in dishes:
    print(f"washing {dish}")
    print(f"{dish} has been cleaned!")
```

### Breakdown
When your computer runs this code, the computer essentially starts by setting `dish` equal to the first element of `dishes` and running through the indented code. Then, it sets `dish` equal to the second element of `dishes` and going through the indented block of code. And then the same for the third, and so on and so on FOR EACH element of `dishes`.

This is all there is to for each loops. Think of it as a function that accepts every element of a list as an input one by one, and then is done.

## For loops
Often times, we want to loop through a set of sequential numbers. This is hardly intuitive, but it actually comes up with surprising frequency. Interestingly enough, because of this, most programming languages' for loops actually loop through sequential numbers instead of looping through each item of a list.

### Range
To make this process easier, python has a built in function called `range()` which can accept a single number as an input, and returns that many numbers, starting with zero, and ending with the number imediately before the input number.

```python
for i in range(5):
    print(i)
```

### Len
Another useful builtin function that one sees in for loops often is the `len()` function which can take a list as an input and returns the number of items in the list.

```python
len([0,0,0,0,0,0])
```

### Eldritch horror
Believe it or not, one often sees both `range()` and `len()` in for loop calls. For example,
if we wanted to make a loop which printed off each element of a list, along with its index, we would have the following:

```python
our_list = ["thing", "thang", "butts"]

for i in range(len(our_list)):
    print(f"index: {i}, value: {our_list[i]}")
```

# While Loops
While loops are a little confusing in my opinion. A while loop is just a conditional statement that keeps repeating until the conditional statement is false. Similarly to a normal if statement, the syntax is `while <conditional>:` and an indented block of code which will be repeated until the conditional statement is false.

Below is a simple enough example of a while loop which will double a number until it's larger than 100.

```python
num = 1

while num < 100:
    num *= 2
    print(num)
```

## Loop Commands
There are two special commands that one can use in loops: break and continue. They both have their uses, and are actually pretty simple to understand.



### Continue

"continue" simply starts the next iteration of a loop, skipping over any code in the remainder of indented block.

```python
words = "this sentence has a lot of e's"

for i in words:
  if i != "e":
    continue
  print(i)
```

### Break

On the other hand, a break statement stops the entire loop in its tracks. It doesn't finish the rest of the current iteration, and it doesn't move on to any further iterations.

```python
i = 0

while True:
  i += 5
  if i > 100:
    break

print(i)
```
