### Division, the tricky one

Because division can be carried out on ints or floats, it behaves in a bit more tricky way.
- If you divide using "/" then the answer is always a float, and given as precisely as possible (for complicated reasons there’s often some error)
- If you divide using "//" then the answer is always an int, and is rounded down. Think of it as doing division and then ignoring the remainder.


```python
4.5 / 0.5
```

```python
38 / 72
```

```python
15 // 14
```

```python
3.8 // 2.2
```

```python
420.69 // 420
```

```python
420.69 / 420
```

### Remainder
One weird operation that is randomly super important when working with computers is the remainder operation. It is done by using the symbol “%” and is essentially the opposite of the “//” operation. It divides the number on the left by the number on the right, and essentially ignores how many times it divides into it and returns just the remainder.

```python
55 % 52
```

```python
77.0 % 2.0
```

## Math with Variables

All of these things work with variables, by which I mean that you can do any of these math operations by using variables in place of actual values.

```python
x = 57
y = 17
x / 7
```

```python
word1 = 'concat'
word2 = 'enate'
word1 + word2
```

As a matter of fact, one can even use mathematical expressions to set the value of a variable.

```python
z = x + y
print(z)
```

### Modifying itself

Now to get freaky. In coding, it's often useful to set the value of a variable to something that is proportionate to its original self. For example, we might want a variable to be set equal to half of its original value.

To enable this, one can use the original value of a variable on the right-hand side of the equals sign to change the value of a variable.

```python
num_infected = 20
num_infected = num_infected + 1
```

```python
y = 30
print(y)
y = y - 1
print(y)
```

### Special support for changing itself

As a matter of fact, these kinds of operations come up frequently enough that there is a shorthand way of performing these operations.

One simply writes a math operator combined with (no spaces between) an equals sign. Then you put the part of the math operation that isn't the original value on the right.

For example if `z` has the value of `4` and we wanted to set `z` subtract `1` from `z` and set `z` equal to the result, we would write `z -= 1`.

```python
z = 4
print(z)

print(z)
```

This kind of math is supported for `+=`, `-=`, `*=`, `/=`, `//=`, `**=`, and `%=`. It's rarely used in practice other than for `+=` though.

## A Final Note

One useful thing to remember is that math operations are, for the most part not restricted to being performed one at a time. You can add 7 things together at once, or even perform multiple different math operations in one line.

When doing this, you can use parentheses just like with normal math, and python knows and follows the sacred ancient ways of PEMDAS.

```python
1 + 2 + 3 + 4 + 5 + 6 + 7
```

```python
4 * (3-1) + 84 // 3
```

```python
1 - (2 - (3 - (4 - 5)))
```

Do note however that sometimes, in the real world, one uses "[]" or "{}" when there are several layers of parentheses and this will NOT work with python. Just use multiple sets of "()" because "[]" and "{}" mean their own special things in python.
