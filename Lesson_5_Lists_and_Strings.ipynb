## Lists

There's one type that gets used all the time which we haven't covered yet: lists.

A list is enclosed in [  ] and contains, as the name would imply, a list of different things separated by commas. In python, this is the default way of storing information that consists of multiple separate entities, like a set of all of the different names of the members in a family, or the different available pokemon types, or anything else that might be inconvenient to divide up into a series of different variables.

```python
numList = [1, 2, 3, 4]
print(numList)
print(1, ",",2, ",",3, ",",4)
```

```python
wordList = ["thing", "thang", "thing"]
print(wordList)
```

Unlike other coding languages, and rather chaotically, lists in python can contain items of different types.

```python
craziness = [1, None, False, "blech this is lawless", 5.340598]
```

Lists can even be empty

```python
empty = []
print(empty)
```

### Referencing List Items

As you can imagine, lists would be worthless if you could only use them as the whole set of values all at once. In most all uses of lists, you typically want to use just one of the values at a time. To reference an individual item from a list, you write the name of the list followed by the number of the item that you want to reference in brackets. Since it's computery stuff, the list numbering starts with 0, meaning that the number for each element is one less than you'd expect.

For example, to get the third element of a list named "myList" one would put "myList[2]"

```python
listy = [1, 2, 3, 4]

print(listy[0])
print(listy[3])
```

## Backwards Referencing

One can also weirdly enough reference elements using negative numbers. When doing this the list references from the back with the final element being -1 and successive bigger negative numbers accessing values further and further from the back of the list.

So, using our list above, "listy[-2]" would return "3"

```python
print(listy[-1])
print(listy[-3])
```

### Lists from within Lists
What if you wanted to reference multiple items from within a list as their own, smaller list? In coding nomenclature, this is referred to as a "slice". In python, this is created using a colon within the brackets.

The format for python is such that if you have one number in the brackets, it references just that item. If you put two numbers (separated with a colon) then it references the sublist starting with the item referenced with the first number and stopping at the item imediately BEFORE the item referenced with the second number.


```python
example = [1, 2, 3, 4, 5, 6, 7, 8]
```

In the above 8 element list, the slice below gets the numbers 2-6.

```python
print(example[1:6])
```

Additionally, if you use just a colon before a number in brackets, it gets a list of all the elements leading up to the item referenced with that number.

And if you use a colon after a number it gets a list of all the elements starting with and after the item referenced with that number.

```python
print(example[:5]) # this prints the numbers 1-5
print(example[3:]) # this prints the numbers 3-8
```

### Editing
One can replace an element of a list by referencing it and using the assignment operator.

```python
# example = [1, 2, 3, 4, 5, 6, 7, 8]
```

```python
example[2] = 3942

print(example)
```

### Inserting

To add an element without replacing an existing one, one uses the name of the list followed with `.insert(<index>, <item>)` where the index is the spot you want the new element to go and the item is what you want to add in. In this case, the elements starting with the index where you're adding an element in will be shifted back one.

Optionally, one can tack a new element on to a list by using the name of a list followed with `.append(<item>)`.

```python
thing = [1,3,4,5,6,7,8,9]

thing.insert(1,2)
print(thing)

thing.append(10)
print(thing)
```

### Pop

To remove an item, one can simply use the name of a list followed with `pop(<index>)` where index is the index number of the element you'd like to remove.

You can also leave the parentheses blank and this will remove the final element.

```python
example_list = [1,2,3,4,5,6,7,8]
print(example_list)

example_list.pop(5)
print(example_list)

example_list.pop()
print(example_list)
```

### Size
Often in coding, you want to find out how many elements are in a particular list. You can get the size of a list by using the `len()` function and put the list inside of the parenthese (just like how you get the type of a variable with the `type()` function).

```python
print(len(example_list))
```

## String Fun

### Substrings

The reason why I lumped strings and lists together is that strings have similar functionality, since strings are, in essence, just lists of characters.

Most importantly, one can use the indices in brackets at the end of a string variable to get a substring or particular letter the same way you get an item or subset of a list.

```python
word = "this is a word"
print(word[:4])
print(word[5:9])
print(word[10:])
```

### Escape Characters

Now to switch directions.

We mentioned earlier how you can use single quotes within a string identified with double quotes or double quotes within a string that is formed using single quotees. You may have wondered however if there is a way to put single or double quotes in a string using the same type of quotation marks. There's a convenient way of inserting special characters into a string, called escape characters. In order to do this, one prefaces the character with a "\".

```python
print("this has \"all sorts\" of \"quotation\" marks in it")
```

You can also use escape characters to put in tabs "\t", a newline "\n", or even a backspace "\b". The one drawback of this method is that if you want to put a backslash you have to type "\\\\".

### f strings

Imagine for a second that you had a program that asked for someone's name and then wanted to print out "you are _ years old". Then, you would need to take the number that is their age, convert it from an int to a string, and concatenate it with the rest of the sentence. Orrrrrrrrrrr...

You could use f strings. An f string is simply a string with an "f" placed at the beginning. This lets you place in special values and variables, enclosed by braces.

For example, the following would create our desired string:

```python
age = 24
our_string = f"You are {age} years old"

print(our_string)
```

### Multiline Strings

This one is quite simple: Python uses new lines to syntactically, but you can make a string that carries across multiple lines by enclosing it in triple double quotes.

```python
big_string = """
This is
a
really
big
long
string
"""

print(big_string)
```

You can also stretch a string a single line by ending it with a single "\\". This however treats the line as if it never ended.

```python
test_num = "thing tha\
t carries onto\
 another line"

print(test_num)
```
